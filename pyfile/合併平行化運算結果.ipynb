```python
import pandas as pd
import time
import os
#"1.exia 2.dynames 3.kyrios 4.virtue"
flist = os.listdir()
path = {'exia':0,'dynames':0,'kyrios':0,'virtue':0}
for f in flist:
    for p in path.keys():
        if p in f:
            path[p] = f
print(path)
df1 = pd.read_csv(path['exia'])
df2 = pd.read_csv(path['dynames'])
df3 = pd.read_csv(path['kyrios'])
df4 = pd.read_csv(path['virtue'])
```

```python
df_merged = df1.append(df2).append(df3).append(df4)
df_merged = df_merged.reset_index(drop=True)
df_merged
```

```python
df_官方 = pd.read_csv('../data/需預測的顧客名單及提交檔案範例.csv')
df_官方
```

```python
def check(df,df2):
    assert df.shape == df2.shape
    assert df.isnull().sum().sum() == 0
    assert (df.dtypes == df2.dtypes).all
    print('test ok')
```

```python
check(df_merged,df_官方)
```

```python
df_merged.to_csv(f'df_knn_method_合併平行化運算結果_{time.time()}.csv',index=False)
```
