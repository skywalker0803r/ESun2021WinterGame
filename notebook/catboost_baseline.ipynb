```python
import pandas as pd

train = pd.read_feather('../data/train_data.feather')
test = pd.read_feather('../data/test_data.feather')
display(train.head())
display(test.head())
```

```python
features = set(train.columns) - set(['index','chid','shop_tag'])
features = list(features)
categorical_features = ['masts','educd','trdtp','naty','poscd','cuorg','gender_code','age','primary_card']
y_name = 'shop_tag'
```

# 切分 train 跟 val

```python
from sklearn.model_selection import KFold
num_Fold = 0
random_seed = 42
kf = KFold(n_splits = 5 ,random_state=0)
Folds = {}
for i,train_val_index in enumerate(kf.split(train.index)):
  Folds[i] = train_val_index
train_index ,val_index = Folds[num_Fold]
val = train.loc[val_index,features+[y_name]]
train = train.loc[train_index,features+[y_name]]
```

# 特徵預處理

```python
# 建立類別特徵索引
categorical_features_index = []
for i in categorical_features:
    categorical_features_index.append(int(features.index(i)))

from tqdm import tqdm
for feature in tqdm(categorical_features):
    train[feature] = train[feature].apply(lambda x:str(x))
    val[feature] = val[feature].apply(lambda x:str(x))
    test[feature] = test[feature].apply(lambda x:str(x))
print(categorical_features_index)
display(test.head())
```

# 定義模型

```python
from catboost import CatBoostClassifier
model = CatBoostClassifier(
    iterations = 100,
    loss_function = 'MultiClass',
    #eval_metric = 'F1',
    #use_best_model = True,
    random_seed = random_seed,
    task_type="CPU", #GPU
    )
```

```python
import gc
gc.collect()
```

# 訓練

```python
test_mode = True

if test_mode == False:
    model.fit(
        train[features],train[y_name],
        eval_set = (val[features],val[y_name]),
        cat_features = categorical_features_index,
        )
if test_mode == True:
    model.fit(
        train[features].head(300).values,train[y_name].head(300),
        cat_features = categorical_features_index,
        )
```

# plot train history

```python
plt.plot(model.evals_result_['learn']['F1'],label='train')
if test_mode == False:
    plt.plot(model.evals_result_['validation']['F1'],label='val')
plt.legend()
plt.show()
```

# F1 report

```python
from sklearn.metrics import classification_report 

print(classification_report(val[y_name],pd.Series(model.predict(val[features])),target_names=sorted(train[y_name].unique())))
```
