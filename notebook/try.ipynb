```python
import pandas as pd
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
import numpy as np
from tqdm import tqdm
import gc
gc.collect()
df = pd.read_feather('../data/2021玉山人工智慧公開挑戰賽冬季賽訓練資料集.feather')
print(df.shape)
df.head()
```

```python
df['shop_tag'].unique()
```

```python
df['shop_tag'] = df['shop_tag'].replace('other','50')
df['shop_tag'].unique()
```

```python
df.columns
```

```python
df['dt'].min(),df['dt'].max()
```

```python
chids = np.random.choice(df['chid'].unique(),size=9999)
chids[0]
```

```python
other_features = ['masts', #婚姻
'educd', #教育
'trdtp', #行業別
'naty', # 國籍
'poscd', # 職位別
'cuorg', #客戶來源
'slam', #正卡信用額度
'gender_code', #信別
'age', #年紀
'primary_card', #正附卡註記
] 
```

```python
sub_df = df.loc[df['chid'].isin(chids)]
sub_df
```

```python
test_data = sub_df.loc[sub_df.dt==24]
test_data
```

```python
train_data = sub_df.loc[sub_df.dt<24]
train_data
```

```python
len(set(test_data.chid)-set(train_data.chid))
```

```python
useful_col = ['domestic_online_cnt','slam','domestic_offline_amt_pct','txn_cnt','txn_amt','dt','domestic_offline_cnt','shop_tag','domestic_online_amt_pct']
y_name = 'shop_tag'
x_col = list(set(useful_col)-set([y_name]))
look_back = 10
batch_X_train = []
batch_X_test = []
batch_y_train = []
batch_y_test = [] 
for chid in tqdm(chids):
    a = train_data[train_data.chid==chid].reset_index(drop=True).drop_duplicates(subset=['dt'])
    
    # split data
    train_df = a.loc[a.dt<a.dt.max()]
    test_df = a.loc[a.dt>a.dt.min()]
    if (len(train_df)==0) | (len(test_df)==0):
        continue
    
    # X,y split
    X_train = train_df.iloc[-1-look_back:-1,:].dropna(axis=0)
    y_train = train_df.iloc[[-1],:].dropna(axis=0)
    X_test = test_df.iloc[-1-look_back:-1,:].dropna(axis=0)
    y_test = test_df.iloc[[-1],:].dropna(axis=0)

    if (len(X_train)!=look_back)|(len(X_test)!=look_back):
        continue
    if (len(y_train)!=1)|(len(y_test)!=1):
        continue

    # reshape
    X_train = X_train[x_col+['shop_tag']].values.reshape(1,-1)
    X_test = X_test[x_col+['shop_tag']].values.reshape(1,-1)
    y_train = y_train[y_name].values
    y_test = y_test[y_name].values
    
    if (X_train.shape[1] != look_back*len(x_col+['shop_tag'])) | (X_test.shape[1] != look_back*len(x_col+['shop_tag'])):
        continue
    
    # append to batch
    batch_X_train.append(X_train)
    batch_X_test.append(X_test)
    batch_y_train.append(y_train)
    batch_y_test.append(y_test)

X_train = np.array(batch_X_train)
X_test = np.array(batch_X_test)
y_train = np.array(batch_y_train)
y_test = np.array(batch_y_test)

print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
b_size = X_train.shape[0]
```

```python
X_train = X_train.reshape(b_size,-1)
X_test = X_test.reshape(b_size,-1)
y_train = y_train.reshape(-1)
y_test = y_test.reshape(-1)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
cat_idx = []
for i,j in enumerate(X_train[0]):
    if type(j) == type('str'):
        cat_idx.append(i)
cat_idx
```

```python
cat_dim = [49 for _ in range(len(cat_idx))]
cat_dim
```

```python
a = pd.DataFrame(X_train)
b = pd.DataFrame(X_test)
for i in tqdm(cat_idx):
    a.iloc[:,i] = pd.to_numeric(a.iloc[:,i])
    b.iloc[:,i] = pd.to_numeric(a.iloc[:,i])
a.shape,b.shape
```

```python
X_train = a.values
X_test = b.values
```

```python
s = np.random.randint(0,len(X_test)-500)
s
```

```python
from catboost import CatBoostClassifier
model = CatBoostClassifier(loss_function='MultiClass',iterations=1000,task_type="CPU")
model.fit(X_train,y_train,cat_features=cat_idx,use_best_model=True,eval_set=(X_test[s:s+500],y_test[s:s+500]),early_stopping_rounds=10)
```

```python
from sklearn.metrics import accuracy_score
X_valid = X_test
y_valid = y_test
preds_mapper = { idx : class_name for idx, class_name in enumerate(model.classes_)}

def get_pred(x):
    preds = model.predict_proba(x)
    y_pred = np.vectorize(preds_mapper.get)(np.argmax(preds, axis=1))
    return y_pred

val_pred = get_pred(X_valid)
train_pred = get_pred(X_train)
valid_acc = accuracy_score(y_pred=val_pred, y_true=y_valid)
train_acc = accuracy_score(y_pred=train_pred, y_true=y_train)
print(f"FINAL TRAIN SCORE FOR : {train_acc}")
print(f"FINAL TEST SCORE FOR : {valid_acc}")
```
