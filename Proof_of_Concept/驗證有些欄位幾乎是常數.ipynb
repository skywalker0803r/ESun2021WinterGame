```python
import pandas as pd
import numpy as np
from tqdm import tqdm
df = pd.read_feather('rank_data.feather')
df
```

```python
col = set(df.columns) - set(df.columns[df.columns.str.contains('shop_tag')]) - set(df.columns[df.columns.str.contains('top')]) - set(['index','dt'])
col = list(col)
df[col]
```

```python
# 隨機取一個chid看他這幾個欄位
chid = np.random.choice(df['chid'].values)
df.loc[df.chid==chid,col]
```

```python
lst = []
# 進行1000次上面cell的實驗,並且計算是否這幾個欄位都是常數(std=0)
for _ in tqdm(range(1000)):
    chid = np.random.choice(df['chid'].values)
    results = np.allclose(df.loc[df.chid==chid,col].describe().loc['std'].sum(),0)
    lst.append(results)
lst
```

```python
np.sum(lst)
```
