```python
import pandas as pd
import numpy as np
from tqdm import tqdm
df = pd.read_feather('../data/2021玉山人工智慧公開挑戰賽冬季賽訓練資料集.feather')
df = df.loc[df.dt>=12,:].dropna(axis=0) # 取最近一年數據即可
print(df.shape)
df
```

```python
import gc
gc.collect()
a = set(df.columns)
b = set(df.columns[df.columns.str.contains('shop_tag')])
c = set(df.columns[df.columns.str.contains('top')])
d = set(['index','dt'])
e = set(df.columns[df.columns.str.contains('pct')])
f = set(df.columns[df.columns.str.contains('cnt')])
g = set(df.columns[df.columns.str.contains('amt')])
h = set(['slam'])
col =  a - b - c - d - e - f - g - h
col = list(col)
df[col]
```

```python
# 隨機取一個chid看他這幾個欄位
chid = np.random.choice(df['chid'].values)
df.loc[df.chid==chid,['dt']+col]
```

```python
df.loc[df.chid==chid,col].describe().loc['std']
```

```python
lst = []
# 進行1000次上面cell的實驗,並且計算是否這幾個欄位都是常數(std=0)
for _ in tqdm(range(1000)):
    chid = np.random.choice(df['chid'].values)
    results = np.allclose(df.loc[df.chid==chid,col].describe().loc['std'].sum(),0)
    lst.append(results)
lst
```

```python
np.sum(lst)
```

```python
print(col)
print('都是常數比例:',np.sum(lst)/len(lst))
```
